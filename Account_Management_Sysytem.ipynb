```python
import sys
import warnings

!pip install art
from art import *

def total_sales(morning_sales, evening_sales):

  '''Calculate total sales for the day'''

  total_sales_per_day = sum(morning_sales) + sum(evening_sales)
  return total_sales_per_day


def workers_salary(hourly_rate, hours_worked):

    '''Calculate worker's salary based on hourly rate and hours worked'''

    worker_salary = hourly_rate * hours_worked
    return worker_salary


def product_profit(total_sales, total_cost):

    '''Calculate the profit of each product (or loss if negative)'''

    profit = total_sales - total_cost
    return profit


def tips_per_shift(shift_sales):

    '''Assuming tips are 2% of shift sales'''

    percentage_of_tips = 0.02
    shift_tips = sum(shift_sales) * percentage_of_tips
    return shift_tips


def total_tips_per_day(morning_sales, evening_sales):

    '''Calculate tips for morning and evening shifts'''

    morning_shift_tips = tips_per_shift(morning_sales)
    evening_shift_tips = tips_per_shift(evening_sales)

    # Calculate total tips for a day
    total_daily_tips = morning_shift_tips + evening_shift_tips
    return total_daily_tips


def display_menu():
  '''
  Program to Automate Martins Store Accounting System
  '''

  print("\nMenu:")
  print("1. Calculate Total Sales for the Day")
  print("2. Calculate Worker's Salary")
  print("3. Calculate Product Profit")
  print("4. Calculate Tips for a Shift")
  print("5. Calculate Total Tips for the Day")
  print("6. Exit")


def get_user_input(prompt):
  '''
  Try-Except to catch any error in the program
  '''
  while True:
      try:
          user_input = float(input(prompt))
          return user_input
      except ValueError:
          print("Invalid input. Please enter a valid number.")


def main():

  '''
    Assuming there is only 3 products in store
    Enter the sales for morning and evening
  '''

  Art=text2art("Martins " + " Store  " + "Account")
  print(Art)
  morning_sales = [get_user_input(f"Enter morning sales for product {i+1}: ") for i in range(2)]  # 3 Products for morning sales
  evening_sales = [get_user_input(f"Enter evening sales for product {i+1}: ") for i in range(2)]  # 3 Products for morning sales
  total_cost = get_user_input("Enter total cost of items sold: ")
  print(f'The total cost of all the sold product is: {total_cost}')

  # To calculate workers rate to know how much to pay
  hourly_rate = get_user_input("Enter hourly rate for the worker: ")
  hours_worked = get_user_input("Enter hours worked by the worker: ")
  print(f'You pay {hourly_rate} hourly and the worker worked for {hours_worked} hours')

  while True:
      display_menu()
      choice = get_user_input("Enter your choice (1-6): ")

      if choice == 1:
          total_sales_for_day_result = total_sales(morning_sales, evening_sales)
          print(f'Total Sales for the Day: ${total_sales_for_day_result:.2f}') # 2 significant fig is added to depict pence (Nigerian Kobo) unit
      elif choice == 2:
          worker_salary_result = workers_salary(hourly_rate, hours_worked)
          print(f'Worker\'s Salary: ${worker_salary_result:.2f}') # 2 significant fig is added to depict pence (Nigerian Kobo) unit
      elif choice == 3:
          profit_result = product_profit(sum(morning_sales) + sum(evening_sales), total_cost)
          print(f'Profit: ${profit_result:.2f}') # 2 significant fig is added to depict pence (Nigerian Kobo) unit
      elif choice == 4:
          shift_choice = get_user_input("Enter 1 for morning shift or 2 for evening shift: ")
          if shift_choice == 1:
              tips_for_shift_result = tips_per_shift(morning_sales)
          elif shift_choice == 2:
              tips_for_shift_result = tips_per_shift(evening_sales)
          else:
              print("Invalid shift choice.")
              continue
          print(f'Tips for the Shift: ${tips_for_shift_result:.2f}') # 2 significant fig is added to depict pence (Nigerian Kobo) unit
      elif choice == 5:
          total_tips_for_day_result = total_tips_per_day(morning_sales, evening_sales)
          print(f'Total Tips for the Day: ${total_tips_for_day_result:.2f}') # 2 significant fig is added to depict pence (Nigerian Kobo) unit
      elif choice == 6:
          print("Exiting the program.... Goodbye! Have a nice day")
          warnings.filterwarnings("ignore")
          SystemExit
          break
      else:
          print("Invalid choice. Please enter a number between 1 and 6.")


if __name__ == '__main__':
  main()
```
